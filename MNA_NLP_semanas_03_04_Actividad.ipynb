**En esta etapa deber谩s llevar a cabo el procesamiento, limpieza y tokenizaci贸n de los comentarios de texto.**

Deber谩s de incluir los procesos que se indicar谩n a continuaci贸n, aunque no necesariament deben ser en este orden. De hecho deber谩s estar revisando la salida de algunos de los registros para que determines cual consideras podr铆a ser el mejor orden de estos pasos.

La incluiremos en una funci贸n llamada "clean_tok()" ya que es la tokenizaci贸n y limpieza b谩sica de cada uno de los 3000 comentarios. La entrada de esta funci贸n son cada uno de estos 3000 strings/comentarios.

Lo que nos regresa "clean_tok()" es una lista de tokens/strings ya simplificados de cada comentario de entrada.

Los pasos a incluir son los siguientes (repito, t煤 deber谩s determinar el orden de ellos, aqu铆 se incluyen en orden alfab茅tico):

*   **Caracteres alfab茅ticos:** Solo considerar caracteres alfab茅ticos. Por lo tanto, deber谩n ser eliminados signos de puntuaci贸n, caracteres especiales y n煤meros.

*   **Longitud mayor a 1:** Se deben considerar solamente tokens de longitud mayor a 1.

*   **Min煤sculas:** Simplificar todo a min煤sculas.

*   **Stopwords:** Eliminar las stopwords.

*   **Tokenizaci贸n:** Cada comentario queda tokenizado por palabras, o cadena de caracteres alfab茅ticos.

Por ejemplo, si el primer comentario de X fuera el string 'All you need is love!', lo que nos regresar铆a clean_tok() ser铆a la lista de strings ['need', 'love'] que llamaremos "tokens".

```
def clean_tok(doc):
  ##############################################################################
  # AGREGA AQU TUS LNEAS DE CDIGO - Pregunta 4:

  tokens = [cadena.lower() for cadena in word_tokenize(re.sub(r'[^a-zA-Z]', ' ', doc)) \
            if len(cadena) > 1 and cadena not in mystopwords]

  # FIN PARA AGREGAR TUS LNEAS DE CDIGO.
  ##############################################################################

  return tokens
```

# üìùEntendiendo el c√≥digo

* **`SparkSession`** : Crea una sesi√≥n de Spark con las siguientes caracter√≠sticas principales.

  * **`local[*]`** : Se especifica que la ejecuci√≥n se realizar√° en modo local.

  * **`appName("PromedioEdades")`** : Forma parte de la configuraci√≥n de una sesi√≥n de Spark al utilizar el constructor. Este m√©todo define un nombre para la aplicaci√≥n de Spark que est√°s creando; para este ejemplo, se define el nombre 'PromedioEdades'.

  * **`getOrCreate()`** : Devuelve una sesi√≥n de Spark existente, si ya ha sido creada previamente, o crea una nueva sesi√≥n en caso contrario.

* **`spark.createDataFrame(data, columns)`** : Crea un DataFrame en PySpark utilizando datos y nombres de columnas definidos.

* **`selectExpr("avg(Edad) as PromedioEdad")`** : Utiliza la funci√≥n `selectExpr` para aplicar una expresi√≥n SQL que calcula el promedio de la columna `Edad` , renombrando el resultado como `PromedioEdad`.

* **`show()`** : Imprime los resultados del DataFrame obtenido tras aplicar la expresi√≥n `avg(Edad)`.

* **`stop()`** : Finaliza la sesi√≥n de Spark y libera los recursos

```python
spark.stop()

# Crear una sesi√≥n de Spark
spark = SparkSession.builder.master("local[*]").appName("PromedioEdades").getOrCreate()

# Crear un DataFrame con datos de ejemplo
data = [("Juan", 25), ("Ana", 35), ("Luis", 42), ("Mar√≠a", 29)]
columns = ["Nombre", "Edad"]
df = spark.createDataFrame(data, columns)

# Mostrar los datos originales
print("Datos originales:")
df.show()

promedio_edad = df.selectExpr("avg(Edad) as PromedioEdad")

# Mostrar el resultado
print("Promedio de edades:")
promedio_edad.show()

# Detener la sesi√≥n de Spark
spark.stop()

```
